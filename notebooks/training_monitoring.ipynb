```python
import pandas as pd
```

```python
df = pd.read_csv("../data/train_data/train_log_unet.csv")

df_filtered = df[df["epoch"].notna()]

print(df_filtered)
```

```python
import matplotlib.pyplot as plt

epochs = df_filtered["epoch"].drop_duplicates().values
train_loss = df_filtered["train/epoch_loss"].drop_duplicates().values
val_loss = df_filtered["val/epoch_loss"].drop_duplicates().values

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(epochs, train_loss, marker="o", label="train/epoch_loss")
ax.plot(epochs, val_loss, marker="D", linestyle="--", label="val/epoch_loss")

ax.set_xlabel("Epoch")
ax.set_ylabel("Loss")
ax.legend()
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("../data/train_data/train_log_unet.csv")
df_filtered_1 = df[df["epoch"].notna()]
df = pd.read_csv("../data/train_data/train_log.csv")
df_filtered_2 = df[df["epoch"].notna()]
```

```python
import matplotlib.pyplot as plt


def plot_model(ax, df, label_prefix, color):
    epochs = df["epoch"].drop_duplicates().values
    train_loss = df["train/epoch_loss"].drop_duplicates().values
    val_loss = df["val/epoch_loss"].drop_duplicates().values

    ax.plot(epochs, train_loss, marker="o", color=color, label=f"{label_prefix} Train")
    ax.plot(
        epochs,
        val_loss,
        marker="D",
        linestyle="--",
        color=color,
        label=f"{label_prefix} Val",
    )


fig, ax = plt.subplots(figsize=(8, 4))

plot_model(ax, df_filtered_1, "Unet", "blue")
plot_model(ax, df_filtered_2, "MLP", "red")

ax.set_xlabel("Epoch")
ax.set_ylabel("Loss")
ax.legend()

plt.tight_layout()
plt.show()
```
