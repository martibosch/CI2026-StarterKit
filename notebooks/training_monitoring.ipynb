```python
import pandas as pd
```

```python
df = pd.read_csv("../data/train_data/train_log.csv")

df_filtered = df[df["epoch"].notna()]

print(df_filtered)
```

```python
import matplotlib.pyplot as plt

epochs     = df_filtered["epoch"].drop_duplicates().values
train_loss = df_filtered["train/epoch_loss"].drop_duplicates().values
val_loss   = df_filtered["val/epoch_loss"].drop_duplicates().values

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(epochs, train_loss, marker="o", label="train/epoch_loss")
ax.plot(epochs, val_loss,   marker="D", linestyle="--", label="val/epoch_loss")

ax.set_xlabel("Epoch")
ax.set_ylabel("Loss")
ax.legend()
plt.tight_layout()
plt.show()
```
